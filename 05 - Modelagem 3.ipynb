```python
from pathlib import Path
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    recall_score,
    precision_score,
    f1_score
)
```

Vamos baixar as bases de treino e teste para trabalharmos

```python
pasta_saida = Path("dados/processed/splits")

x_treino_smote = pd.read_csv(pasta_saida / "x_treino_smote.csv")
x_teste = pd.read_csv(pasta_saida / "x_teste.csv")

y_treino_smote = pd.read_csv(pasta_saida / "y_treino_smote.csv")["internacao_prolongada"]
y_teste = pd.read_csv(pasta_saida / "y_teste.csv")["internacao_prolongada"]
```

Criar uma função para salvar os resultados dessa fase

```python
# função para guardar os resultados dessa fase

resultados_final = []

def salvar_resultados_final(nome_modelo, y_real, y_pred): # uma função para facilitar salvar os resultados
    tn, fp, fn, tp = confusion_matrix(y_real, y_pred).ravel()
    
    resultados_final.append({
        'Modelo': nome_modelo,
        'Accuracy': accuracy_score(y_real, y_pred),
        'Recall': recall_score(y_real, y_pred),
        'Precision': precision_score(y_real, y_pred),
        'F1-score': f1_score(y_real, y_pred),
        'Verdadeiros Negativos': tn,
        'Falsos Positivos': fp,
        'Falsos Negativos': fn,
        'Verdadeiros Positivos': tp
    })
```

Apesar da Regressão Logística ajustada ter apresentado melhor equilibrio geral entre recall, precision e f1-score, o XGBoost ajustado foi selecionado para a etapa seguinte por, alem de apresentar recall elevado, ter maior flexibilidade de configuração

O objetivo desta etapa é testar novas combinações de hiperparametros do XGBoost, buscando reduzir falsos positivos e melhorar a precision e o f1-score da classe 1, sem perder demasiadamente a capacidade de identificar internações prolongadas
